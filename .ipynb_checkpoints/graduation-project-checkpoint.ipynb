## Model Eğitimi

```python
model = Sequential()

model.add(Dense(1024, activation="relu", kernel_initializer='glorot_uniform', input_dim=3,))

model.add(Dense(512, activation="relu", kernel_initializer='glorot_uniform'  ))
model.add(Dense(256, activation="relu", kernel_initializer='glorot_uniform' ))
model.add(Dense(256, activation="relu",  kernel_initializer='glorot_uniform'))
model.add(Dense(128, activation="relu",  kernel_initializer='glorot_uniform'))
model.add(Dense(64, activation="relu",  kernel_initializer='glorot_uniform'))



model.add(Dense(1, activation="softplus", kernel_initializer='glorot_uniform'))
```

```python
adam = Adam(learning_rate=0.0001)
Adam = Adam(learning_rate=0.001, decay=1e-3 / 200)
```

```python
sgd = SGD(learning_rate = 0.1, momentum=0.1)
```

```python
rms = RMSprop(learning_rate = 0.0001, momentum = 0.0)
```

```python
model.compile(loss='mse', optimizer = 'adam') 
```

```python
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=200) 
```

```python
print(model.summary())
```

```python
model.fit(x=x_train, y=y_train, epochs=10000, validation_data=(x_test, y_test),  verbose=1,  batch_size=50, callbacks =[es] )
```

```python
model.history.history
```
